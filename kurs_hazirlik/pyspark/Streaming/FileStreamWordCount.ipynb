```python
import findspark
findspark.init()
```

```python
from pyspark.streaming import *
from pyspark import SparkConf, SparkContext
```

```python
sc = SparkContext("local[4]", "FileStreamWordCount")
```

```python
ssc = StreamingContext(sc, batchDuration=10)
```

```python
lines = ssc.textFileStream("D:\\spark-streaming-test")
```

```python
words = lines.flatMap(lambda x: x.split(" "))
```

```python
mapped_words = words.map(lambda x: (x, 1))
```

```python
wordcounts = mapped_words.reduceByKey(lambda x,y: x+y)
```

```python
sorted_words = wordcounts.map(lambda x: (x[1], x[0])).transform(lambda x: x.sortByKey(False))
```

```python
sorted_words.pprint()
```

```python
ssc.start()
```

```python
ssc.awaitTermination()
```

```python
ssc.stop()
```
